# Bi-encoder

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.models import Pooling, Transformer

def get_bi_encoder(bi_encoder_name):
    raw_model = Transformer(model_name_or_path=bi_encoder_name)
    bi_encoder_dim = raw_model.get_word_embedding_dimension()
    
    pooling_model = Pooling(
        bi_encoder_dim,
        pooling_mode_cls_token = False,
        pooling_mode_mean_tokens = True
    )
    bi_encoder = SentenceTransformer(
        modules = [raw_model, pooling_model],
        device = 'cpu' 
    )
    
    return bi_encoder, bi_encoder_dim

# Формируем из строки вектор
def str_to_vec(bi_encoder, text):
    embeddings = bi_encoder.encode(
        text,
        convert_to_tensor=True,
        show_progress_bar=False
    )
    return embeddings.tolist()  # Конвертируем тензор в список
```

# Поиск векторов

```python
from qdrant_client import QdrantClient

# Создаем подключение к векторной БД
qdrant_client = QdrantClient('http://localhost:6333/dashboard', port=6333, timeout=60)


def vec_search(bi_encoder, encoder_name, query, n_top_cos):
    COLL_NAME = encoder_name.replace('/','_')
    
    # Кодируем запрос в вектор
    query_emb = str_to_vec(bi_encoder, query)

    # Поиск в БД
    search_result = qdrant_client.search(
        collection_name = COLL_NAME,
        query_vector = query_emb,
        limit = n_top_cos,
        with_vectors = False
    )
    top_art = [x.payload['description'] for x in search_result]
    top_url = [x.payload['url'] for x in search_result]
    top_title = [x.payload['title'] for x in search_result]
    
    return top_title, top_art, top_url
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = 'IlyaGusev/saiga_llama3_8b'

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map='cpu'
)
```

```python
PROMPT_TEMP = '''<|begin_of_text|><|start_header_id|>system<|end_header_id|>
Ты — Сайга, русскоязычный автоматический ассистент. Ты разговариваешь с людьми и помогаешь им.<|eot_id|><|start_header_id|>user<|end_header_id|>
Используй только следующий контекст, созданный на основе системы faq, чтобы точно ответить на вопрос в конце, который задают пользователи социальной сети ВКонтаке. Тебе не нужно передавать никакие ссылки, напиши только текстовый ответ. 
Контекст:
###
{chunks_join}
###
Вопрос:
###
{query}<|eot_id|><|start_header_id|>assistant<|end_header_id|>'''

def get_llm_answer(query, chunks_join, max_new_tokens, temperature, top_p, top_k):

    prompt = PROMPT_TEMP.format(chunks_join=chunks_join, query=query)
    
    def generate(model, tokenizer, prompt):
        data = tokenizer(prompt, return_tensors="pt", add_special_tokens=False)
        data = {k: v.to(model.device) for k, v in data.items()}
        output_ids = model.generate(
            **data,
            bos_token_id=128000,
            eos_token_id=128009,
            pad_token_id=128000,
            do_sample=True,
            max_new_tokens=max_new_tokens,
            no_repeat_ngram_size=15,
            repetition_penalty=1.12,
            temperature=temperature,
            top_k=top_k,
            top_p=top_p 
        )[0]
        output_ids = output_ids[len(data["input_ids"][0]) :]
        output = tokenizer.decode(output_ids, skip_special_tokens=True)
        return output.strip()
    
    response = generate(model, tokenizer, prompt)
    
    return response
```

# Оценка

```python
import re
import json
from rouge import Rouge
from pymorphy2 import MorphAnalyzer

f = open('stopwords-ru.json', encoding='utf-8')
stop_words = json.load(f)

morph = MorphAnalyzer()
patterns = "[«»°!#$%&'()*+,./:;<=>?@[\]^_`{|}~—\"\-]+"

def lemmatize(string):
    clear = re.sub(patterns, ' ', string)
    tokens = []
    for token in clear.split():
        if token:
            token = token.strip()
            token = morph.normal_forms(token)[0]
            if token not in stop_words:
                tokens.append(token)
    tokens = ' '.join(tokens)
    return tokens

def get_llm_score(answer, answer_true):
    answer = lemmatize(answer)
    answer_true = lemmatize(answer_true)
    if len(answer) == 0:
        answer = 'Ответ не найден'
        
    rouge = Rouge()
    scores = rouge.get_scores(answer, answer_true)[0]
    rouge_1 = scores['rouge-1']['f']
    rouge_2 = scores['rouge-2']['f']
    
    return rouge_1, rouge_2
```

# Выполнение теста

```python
#from tqdm.auto import tqdm
#Заголовок: 
text_temp = '''{title}
{description}
'''

def run_one_test(
        df, encoder_name, 
        n_top_cos, join_sym, 
        max_new_tokens, temperature, top_p, top_k):
    try:
        bi_encoder, vec_size = get_bi_encoder(encoder_name)
        print(bi_encoder)
        result = []
        for i, row in q_df.iterrows():
            query = row['Вопрос']
            answer_true = row['Правильный ответ']

            top_title, top_art, top_url = vec_search(bi_encoder, encoder_name, query, n_top_cos)
            row['top_url'] = top_url
            print('search completed')
            chunks = [text_temp.format(title=t, description=a) for t, a in zip(top_title, top_art)]
            chunks_join = join_sym.join(chunks)
            print(f'Вопрос: {query}')  # Выводим вопрос

            answer = get_llm_answer(query, chunks_join, max_new_tokens, temperature, top_p, top_k)
            print('get answer')
            print(answer)
            url = row['url'] if 'url' in row else []
            row['Ответ'] = answer
            row['art_score'] = None if len(url) == 0 else len([u for u in url if u in top_url]) / len(url)
            rouge_1, rouge_2 = get_llm_score(answer, answer_true)
            row['rouge_1'] = rouge_1 * 100
            row['rouge_2'] = rouge_2 * 100
            row['score'] = row[['rouge_1','rouge_2']].mean() # * row['art_score']

            result.append(row)

        result = pd.DataFrame(result)
        result = result.sort_values(by=['score'], ascending=False)
        result = result.reset_index(drop=True)

        score = result['score'].mean()

        return result, score
    
    except Exception as e:
        print(e)
        return None, 0
```

# Запуск Optuna'ы

```python
import optuna

def objective(trial):
    global best_score, best_result
    
    encoder_name = trial.suggest_categorical('encoder_name', ['cointegrated/rubert-tiny2',
                                                              'kazzand/ru-longformer-large-4096',
                                                              'cointegrated/LaBSE-en-ru',
                                                              'intfloat/e5-large-v2',
                                                              'intfloat/multilingual-e5-large',
                                                              'sentence-transformers/paraphrase-multilingual-mpnet-base-v2',
                                                              'sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2' 
                                                             ])
    join_sym = trial.suggest_categorical('join_sym', [' ',' | ','\n','\n\n','\n\n\n','\n---\n','\n===\n'])
    
    #no_repeat_ngram_size = trial.suggest_int('no_repeat_ngram_size', 1, 20)
    #repetition_penalty  = trial.suggest_float('repetition_penalty ', 1.0, 1.3)
    
    n_top_cos = trial.suggest_int('n_top_cos', 1, 15)
    
    max_new_tokens = trial.suggest_int('max_new_tokens', 50, 1500)
    temperature = trial.suggest_float('temperature', 0.01, 0.99)
    top_p = trial.suggest_float('top_p', 0.01, 0.99)
    top_k = trial.suggest_int('top_k', 10, 150)
    
    result, score = result, score = run_one_test(
        q_df, encoder_name, n_top_cos, join_sym, max_new_tokens, temperature, top_p, top_k)
    
    if score > best_score:
        best_score = score
        best_score_tag = ' <--'
        best_result = result
    else:
        best_score_tag = ''

    print(f'{score:.2f}', best_score_tag)
    #print(f'- encoder_name: {encoder_name}')
    #print(f'- sep: {sep} | chunk_size: {chunk_size} | chunk_overlap: {chunk_overlap} | top_cos: {top_cos}')
    #print(f'- max_new_tokens: {max_new_tokens} | temperature: {temperature} | top_p: {top_p} | top_k: {top_k}')
    
    return score

optuna.logging.set_verbosity(optuna.logging.WARNING) 
```

```python
import pandas as pd

best_score = 0
best_result = None

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=5000, timeout=60*60*7)
```

# Qdrant

```python
from qdrant_client import QdrantClient
from qdrant_client.http.models import Distance, VectorParams, PointStruct
import uuid

# Создаем подключение к векторной БД
#qdrant_client = QdrantClient(url='http://localhost', port=6333, timeout=60)


#Заголовок: 
text_temp = '''{title}
{description}
'''

def art_to_vecdb(bi_encoders_name): # sep, chunk_size, chunk_overlap
    COLL_NAME = bi_encoders_name.replace('/','_')
    print(COLL_NAME)
    bi_encoder, vec_size = get_bi_encoder(bi_encoders_name)
    
    # Удаляем и заново создаем коллекцию
    qdrant_client.delete_collection(collection_name=COLL_NAME)
    qdrant_client.create_collection(
        collection_name = COLL_NAME,
        vectors_config = VectorParams(size=vec_size, distance=Distance.COSINE),
    )
    
    tdfs = [art_df[:1000], art_df[1000:2000], art_df[2000:4000]]
    
    for t_df in tdfs:
        points = []
        for i, row in t_df.iterrows():
            title = row['title']
            description = row['description']
            url = row['url']

            text = text_temp.format(title=title, description=description)
            emb = str_to_vec(bi_encoder, text) # Convert tensor to list

            point = PointStruct(
                id = str(uuid.uuid4()),
                vector = emb, 
                payload={'title': title, 
                         'description': description, 
                         'url': url, 
                        }
            )
            points.append(point)


        operation_info = qdrant_client.upsert(
            collection_name = COLL_NAME,
            wait = True,
            points = points
        )
```

```python
bi_encoders = ['intfloat/multilingual-e5-large']
                        
for bi_encoders_name in bi_encoders:
    art_to_vecdb(bi_encoders_name)
```
